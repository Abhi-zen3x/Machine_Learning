#**MEAN SHIFTING**

##**IMPORTING MODULE**

```python
from sklearn.datasets import load_iris
from sklearn.cluster import MeanShift,estimate_bandwidth
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

##**DATA LOADING**

```python
data=load_iris()
X,Y=data.data,data.target
```

##**STANDARDIZING DATA**

```python
x=StandardScaler().fit_transform(X)
```

##**ESTIMATING BANDWIDTH**

```python
bandwth=estimate_bandwidth(x,quantile=0.3)
```

##**MODEL FITTING**

```python
model=MeanShift()
model.fit(x)
label=model.labels_
```

##**EVALUATION**

```python
sil=silhouette_score(X,label)
print("SILHOUETTE SCORE:",sil)
```

##**GRAPH PLOTING**

```python
plt.title("Mean shifting")
plt.scatter(X[:,0],X[:,1],c=label,cmap='rainbow',edgecolors='k')
plt.xlabel("features")
plt.ylabel("labels")
plt.legend()
plt.show()
```
