#**REGULARIZATION**

###**IMPORTING MODULE**

```python

import numpy as np
from sklearn.linear_model import RidgeCV
from sklearn.metrics import mean_absolute_error,r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
```

###**LOADING DATA**

```python
x = np.array([[12],[23],[34],[45],[56],[67],[78],[89],[123],[134]])
Y= np.array([240,1135,2568,4521,7865,9236,11932,14589,19856,23145])
```

###**TRAINING MODEL**

```python
model=RidgeCV()
poly=PolynomialFeatures(degree=3)
X=poly.fit_transform(x)
model.fit(X,Y)
```

###**PREDICTING OUTPUT**

```python
Y_predict=model.predict(X)
```

###**EVALUATION**

```python
mae=mean_absolute_error(Y,Y_predict)
r2=r2_score(Y,Y_predict)
print("Mean absolute error:",mae)
print("R2 score:",r2)
```

###**PLOTING GRAPH**

```python

plt.figure(figsize=[5,5])
plt.scatter(x,Y,s=40,c='yellow',label='datapoints',marker='*',edgecolors='green')
plt.plot(x,Y_predict,c='red',label='regression')
plt.xlabel('Feature')
plt.ylabel('label')
plt.title('Lir REg')
plt.legend()
plt.show()
```
