#**LINEAR REGRESSION**

```python

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score
import matplotlib.pyplot as plt
```

```python
mni=LinearRegression()
mni.fit(X,Y)
```
